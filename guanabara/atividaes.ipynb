```python
numero_1 = int(input("Informe o primeiro numero: "))
numero_2 = int(input("Informe o segundo numero: "))

resultado = numero_1 + numero_2
print(f"o resultado de {numero_1} + {numero_2} e igual a {resultado}")
```

```python
palavra = input("Digite a palavra: ")
print(f"O tipo primitivo desse valor e: {type(palavra)}")
print(f"So tem espacos? {palavra.isspace()}")
print(f"E um numero? {palavra.isnumeric()}")
print(f"E alfabetico? {palavra.isalpha()}")
print(f"E alfanumerico? {palavra.isalnum()}")
print(f"Esta em maiusculo? {palavra.isupper()}")
print(f"Esta em minusculo? {palavra.islower()}")
print(f"E um titulo? {palavra.istitle()}")
```
