```python
media = 0
media_requrida = int(input("Media requerida: "))
def dar_parabens(valor_marco):
    print(f"Parabens por alcancar um marco de {valor_marco}".upper())

while media < media_requrida:
    if media != 20 and media != 50:
        print(f"Media atual {media} no entanto a requerida e {media_requrida}")
    media = media + 1
    if media == 20:
        dar_parabens(media)
    elif media == 50:
        dar_parabens(media)

print(f"A media de {media_requrida} foi batida".upper())
```
