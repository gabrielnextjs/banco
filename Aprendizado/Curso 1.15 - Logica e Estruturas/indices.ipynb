```python
try:
    nome = input("Qual e o seu nome? ")
    salario_mensal = float(input(f"Ola {nome} digite seu salario mensal apenas com numeros: "))
    horas_trabalhadas = float(input(f"{nome} Qual e a sua quantiade de horas trabalhadas? "))
    valor_horas = salario_mensal / horas_trabalhadas
    valor_horas = round(valor_horas, 2)
    print(f"R${valor_horas}")

except ValueError:
    print(f"{nome}. O seu salario mensal ou horas trabalhadas devem ser insiridas apenas usando numeros.")
except ZeroDivisionError:
    print(f"{nome}. Voce deve digitar um valor maior que zero")
```

```python
nomes = []
opcao = 0
import time

def adicionar(nome_novo):
    novo_nome = {"nome": nome_novo}
    nomes.append(novo_nome)
try:

    while opcao != 4:
        opcao = int(input('''
    OPCAO [1]: ADICIONAR PESSOA
    OPCAO [2]: EXCLUIR PESSOA
    OPCAO [3]: VER NOMES
    OPCAO [4]: SAIR                                                                       
                        '''))

        if opcao == 4:
            print("Saindo do sistema")
            time.sleep(2)
            break
        elif opcao == 3:
            print(nomes)
        elif opcao == 2:
            excluir = input("Insira o nome que voce deseja excluir: ")
            encontrado = False
            for nome in nomes:
                if nome["nome"] == excluir:
                    nomes.remove(nome)
                    encontrado = True
                    print("Nome removido com sucesso")
                    break
            if not encontrado:
                print("Nome nao encontrado")
        elif opcao == 1:
            nome_adicionar = input("Qual nome voce quer adcionar a lista? ")
            adicionar(nome_adicionar)
        else:
            print("Esta opcao nao existe")



except ValueError:
    print("Digite apenas numeros")
except ZeroDivisionError:
    print("Digite apenas opcoes como [1],[2],[3]")

```
