```python
nome = input("Qual e seu nome? ")
idade = int(input("Qual e sua idade? "))
produto = 0
Produtos = []
idention = False
nome = nome.title()

def comprar(produto):
    while produto != 5:
        if produto == 1:
            Produtos.append(produto)
            print(f"O produto Cerveja foi comprado")
        elif produto == 2:
            Produtos.append(produto)
            print(f"O produto Cigarro foi comprado")
        elif produto == 3:
            Produtos.append(produto)
            print(f"O produto Sabonete foi comprado")
        elif produto == 4:
            Produtos.append(produto)
            print(f"O produto Cogumelo foi comprado")
        elif produto == 5:
            break
        return produto





if idade >= 18:
    print("Acesso liberado")
    idention = True
else:
    print("Acesso negado")


if idention == True:
    produto = int(input("Oque voce deseja comprar? [1]Cerveja [2]Cigarro [3]Sabonete [4]Cogumelo [5]Sair"))
    produto = comprar(produto)
    

    
    




```
