```python
faturamento = 1000
custo = 300
imposto = 0.2
lucro_1 = faturamento - custo - imposto * faturamento


faturamento = 1000
lucro_2 = faturamento - custo - imposto * faturamento

print(f"Lucro 1 foi de: R$:{lucro_1}")
print(f"Lucro 2 foi de: R$:{lucro_2}")


margem_lucro = lucro_2 / faturamento
print(f"Margem de lucro de: R${margem_lucro}")


meta = 0.5

if margem_lucro >= meta:
    print("Bateu a meta")
else:
    print("Meta nao batida")
```
