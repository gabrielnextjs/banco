```python
def calcular__imposto(valor):
    if valor < 1000:
        imposto = valor * 0.1
    elif valor < 2000:
        imposto = valor * 0.13
    else:
        imposto = valor * 0.2

    return imposto

preco_produto1 = 2500
preco_produto2 = 3500

imposto_produto1 = calcular__imposto(preco_produto1)
imposto_produto2 = calcular__imposto(preco_produto2)

print(imposto_produto1,"e",imposto_produto2)
```

```python
def vereficar_par(numero):
    if numero % 2 == 0:
        return True
    else:
        return False

vereficar = int(input("Insira o numero a qual deseja vereficar: "))

if vereficar_par(vereficar):
    print(f"o numero {vereficar} e par")
else:
    print(f"O numero {vereficar} e impar")
    


```

```python

def somar_lista(*numeros):
    resultado = 0
    for numero in numeros:
        resultado += numero
    return resultado

soma_da_lista = somar_lista(6, 2, 7, 9)
print(f"O resultado e {soma_da_lista}")
```

```python
def calcular__media(*numeros):
    qtd = len(numeros)
    soma = 0
    for  numero in numeros:
        soma += numero
    media = soma / qtd
    return media

resultado = calcular__media(7, 2, 4, 9)
print(resultado)
```

```python
def informacoes_pessoais(**informacoes):
    for chave,  valor in informacoes.items():
        print(f"{chave}: {valor}")

nome = input("Insira seu nome ")
idade = int(input("Insira sua idade: "))
estado = input("Insira seu estado: ")

informacoes_pessoais(nome = nome, idade = idade , estado = estado)
```
