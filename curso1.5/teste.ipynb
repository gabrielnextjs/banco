```python
def avaliar_membro(nome,pontuacao):
    if pontuacao >= 1000:
        print(f"O jogador {nome} atingiu {pontuacao} pontos e foi promovido para a Elite da STREET!")
    else:
        print(f"O jogador {nome} tem {pontuacao} pontos e ainda precisa treinar mais.")


print("--- INICIANDO O SISTEMA DA GUILDA ---")

membros = int(input("Em numeros, insira o numero de mebros que iram treinar hoje:"))
for i in range(membros):
    nome = input(f"Insira o nom do membro {i + 1}")
    pontuacao = int(input(f"Em numeros, insira a pontuacao do jogador {nome}"))

    avaliar_membro(nome, pontuacao)
```
